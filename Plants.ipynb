```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Set Seaborn style for plots
sns.set(style="whitegrid")

# Example data
pm25_data = [200, 220, 180, 160, 170, 150, 190]  # Historical PM2.5 data
forecast_steps = 48
plant_data = pd.DataFrame({
    "Location": ["Anand Vihar", "Hauz Khas", "Shahdara", "Rajokari", "Najafgarh"],
    "Soil Type": ["Sand loam", "Alluvial", "Alluvial", "Rocky soil", "Clayey Soil"],
    "Recommended Plants": [
        ["Peepal", "Banyan", "Ashoka", "Mango", "Tulsi"],
        ["Peepal", "Neem", "Arjun", "Ashoka", "Money plant", "Aloe vera"],
        ["Peepal", "Jamun", "Banyan", "Neem", "Mango", "Money plant", "Aloe vera", "Amla shrubs"],
        ["Banyan", "Mango", "Neem", "Peepal", "Gulmohar", "Ashoka", "Money plant", "Amla shrubs", "Snake plant"],
        ["Banyan", "Mango", "Arjun", "Peepal", "Gulmohar", "Bougenvilla Shrubs", "Amla shrubs", "Snake plant", "Hibicus shrubs"]
    ],
    "5 years (%)": [30, 40, 50, 45, 60],
    "10 years (%)": [50, 60, 70, 65, 75]
})

# 1. Forecasting PM2.5 using SARIMAX (Model 1)
def forecast_pm25(data, order=(1, 0, 4), seasonal_order=(1, 1, 0, 24), steps=48):
    model = SARIMAX(data, order=order, seasonal_order=seasonal_order)
    results = model.fit(disp=False)
    forecast = results.get_forecast(steps=steps)
    forecast_values = forecast.predicted_mean
    return forecast_values

forecasted_pm25 = forecast_pm25(pm25_data)

# 2. Generate visualization for the forecasted PM2.5 data
def plot_forecast(historical, forecast, steps):
    plt.figure(figsize=(12, 6))
    plt.plot(range(len(historical)), historical, label="Historical PM2.5", color="blue")
    plt.plot(range(len(historical), len(historical) + steps), forecast, label="Forecasted PM2.5", color="orange")
    plt.axhline(y=100, color='red', linestyle='--', label="Safe PM2.5 Threshold")
    plt.title("PM2.5 Forecast (Model 1)")
    plt.xlabel("Time Steps")
    plt.ylabel("PM2.5 Concentration")
    plt.legend()
    plt.show()

plot_forecast(pm25_data, forecasted_pm25, forecast_steps)

# 3. Pollution Reduction Recommendation (Model 2)
location = "Anand Vihar"
soil_type = "Sand loam"
selected_plants = plant_data.loc[plant_data["Location"] == location, "Recommended Plants"].values[0]
pollution_reduction_5 = plant_data.loc[plant_data["Location"] == location, "5 years (%)"].values[0]
pollution_reduction_10 = plant_data.loc[plant_data["Location"] == location, "10 years (%)"].values[0]

# 4. Visualize Pollution Reduction for Recommended Plants
def plot_pollution_reduction(plants, reduction_5, reduction_10):
    plt.figure(figsize=(12, 6))
    x = np.arange(len(plants))
    width = 0.35
    plt.bar(x - width/2, reduction_5, width, label="Reduction in 5 years", color="green")
    plt.bar(x + width/2, reduction_10, width, label="Reduction in 10 years", color="darkgreen")
    plt.xticks(x, plants, rotation=45)
    plt.title("Pollution Reduction by Recommended Plants (Model 2)")
    plt.ylabel("Reduction Percentage (%)")
    plt.legend()
    plt.show()

plot_pollution_reduction(selected_plants, [pollution_reduction_5], [pollution_reduction_10])

# 5. Pollution Reduction Over Time (Model 3)
import numpy as np

def plot_cumulative_reduction(pm25_data, reduction_percentage):
    time = np.arange(len(pm25_data))
    reduction_factor = reduction_percentage / 100  # Convert percentage to decimal
    cumulative_reduction = pm25_data - np.array(pm25_data) * reduction_factor 
    # Convert pm25_data to a NumPy array for element-wise multiplication

    plt.figure(figsize=(12, 6))
    plt.plot(time, pm25_data, label="Original PM2.5", color="blue")
    plt.plot(time, cumulative_reduction, label="Reduced PM2.5", color="red")
    plt.legend()
    plt.show()

plot_cumulative_reduction(pm25_data, reduction_percentage=40)
```
